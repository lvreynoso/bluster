```python
import networkx as nx
G = nx.Graph()
```

```python
import csv

with open('media_data.csv') as f:
    readCSV = csv.reader(f, delimiter=',')
    next(readCSV)
    for row in readCSV:
        author = row[0]
        outlet = row[1]
        years = row[2]
        if author not in G:
            G.add_node(author)
        if outlet not in G:
            G.add_node(outlet)
        G.add_edge(author, outlet, weight=years)
```

```python
# unweighted plot

import matplotlib.pyplot as plt

plt.figure(figsize=(50, 50))

nx.draw(G, with_labels=True, font_weight='bold')
plt.savefig('unweighted.png')
plt.show()
```
